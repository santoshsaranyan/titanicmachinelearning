```python
d_train=train.drop(columns=['PassengerId','Cabin','Name','Ticket'])
d_test=test.drop(columns=['Name','PassengerId','Cabin','Ticket'])
d_train=d_train.dropna(subset=['Embarked'])
```

```python
display(d_train.isnull().sum())
display(d_test.isnull().sum())
display(d_test)
```

```python
from sklearn.model_selection import train_test_split
#First split the data into features and target
X=d_train.drop(columns=['Survived'])
y=d_train['Survived']

#Use train_test_split to split the data
train_x,val_x,train_y,val_y=train_test_split(X,y,test_size=0.2,random_state=7)

#Getting the categorical columns
cat_cols=[col for col in train_x.columns if train_x[col].dtype=='object']
```

```python
from sklearn import preprocessing
encoder=preprocessing.LabelEncoder()
#Applying the encoding
for col in cat_cols:
    train_x[col+'_label']=encoder.fit_transform(train_x[col])
    val_x[col+'_label']=encoder.fit_transform(val_x[col])
    d_test[col+'_label']=encoder.fit_transform(d_test[col])
#dropping the categorical columns
train_x=train_x.drop(columns=cat_cols)
val_x=val_x.drop(columns=cat_cols)
d_test=d_test.drop(columns=cat_cols)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer()
imp_train_x= pd.DataFrame(imputer.fit_transform(train_x))
imp_val_x = pd.DataFrame(imputer.transform(val_x))
imp_test= pd.DataFrame(imputer.fit_transform(d_test))

# Imputation removed column names; put them back
imp_train_x.columns = train_x.columns
imp_val_x.columns = val_x.columns
imp_test.columns = d_test.columns
```

```python
display(imp_test)
```

```python
imp_val_x['Family']=imp_val_x['SibSp']+imp_val_x['Parch']
imp_train_x['Family']=imp_train_x['SibSp']+imp_train_x['Parch']
imp_test['Family']=imp_test['SibSp']+imp_test['Parch']
imp_train_x['Alone']=[1 if x==0 else 0 for x in imp_train_x['Family']]
imp_val_x['Alone']=[1 if x==0 else 0 for x in imp_val_x['Family']]
imp_test['Alone']=[1 if x==0 else 0 for x in imp_test['Family']]
display(imp_train_x)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

rand_forest = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=7)
rand_forest.fit(imp_train_x,train_y)
pred=rand_forest.predict(imp_val_x)

val_score=roc_auc_score(val_y,pred)

val_score
```

```python
predictions = rand_forest.predict(imp_test)

output = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': predictions})
output.to_csv('my_submission.csv', index=False)
print("Your submission was successfully saved!")
```
